## 6.3 填充和步幅
每次卷积会使$n_k \times n_w$的输入变为$n_o \times n_h$的输出，其中$n_o = \frac{n_i - n_k + 2p}{s} + 1$。
1. 填充(padding)：填充操作将输入的边界区域扩展为与输入相同大小，从而保证输出的边界区域与输入相同大小。
2. 步幅(stride)：大幅度降低图像的宽度和高度，剔除冗余信息。

### 6.3.1 填充

```python
import torch
from torch import nn

# 为了方便起见，我们定义了一个计算卷积层的函数。
# 此函数初始化卷积层权重，并对输入和输出提高和缩减相应的维数
def comp_conv2d(conv2d, X):
    # 这里的(1, 1)表示批量大小和通道数都是1
    X = X.reshape((1, 1) + X.shape)
    Y = conv2d(X)
    return Y.reshape(Y.shape[2:])

conv2d = nn.Conv2d(1, 1, kernel_size=3, padding=1)
X = torch.rand(size=(8, 8))
comp_conv2d(conv2d, X).shape
```

```python
conv2d = nn.Conv2d(1, 1, kernel_size=(5, 3), padding=(2, 1))
comp_conv2d(conv2d, X).shape
```

### 6.3.2 步幅
我们将每次滑动元素的数量称为步幅。
当垂直步幅为$s_h$时，水平步幅为$s_w$时，输出形状为：
$$
\lfloor \frac{(n_h - k_h + p_h + s_h)}{s_h}\rfloor \times \lfloor \frac{(n_w - k_w + p_w + s_w)}{s_w}\rfloor
$$
进一步如果设置了$p_h = k_h - 1$和$p_w = k_w - 1$，那么输出形状为：
$$
\lfloor \frac{n_h + s_h - 1}{s_h}\rfloor \times \lfloor \frac{n_w + s_w - 1}{s_w}\rfloor
$$
如果输入的高度和宽度可以被垂直和水平步幅整出，则输出形状将为：
$$
\frac{n_h}{s_h} \times \frac{n_w}{s_w}
$$

```python
conv2d = nn.Conv2d(1, 1, kernel_size=3, padding= 1, stride=2)
comp_conv2d(conv2d, X).shape
```

```python
conv2d = nn.Conv2d(1, 1, kernel_size=(3, 5), padding=(0 ,1), stride=(3, 4))
comp_conv2d(conv2d, X).shape
```
